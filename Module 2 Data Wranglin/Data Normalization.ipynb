```python
#For data comparability normalization is quintessential
#Methods of Normalizing Data (Common)
#1.Simple feature scaling=(X_old/X_max) 2.Min-Max=(X_old-X_min)/(X_max-X_min) 3. Z-score=(X_old-Population Avg.)/Standard Deviation
```

```python
#Simple Feature
import pandas as pd
path="C:\\Users\\ARNOB\\Desktop\Automobile.csv"
df=pd.read_csv(path)
df
```

```python
df1=df["length"]/df["length"].max()
```

```python
df1
```

```python
#Min-Max method
df2=(df["length"]-df['length'].min())/(df["length"].max()-df['length'].min())
df2
```

```python
#Z-Score Method THE REAL DEAL
df3=(df['length']-df['length'].mean())/(df['length'].std())
df3
#We should probably wrangle the data by pre-processing and deciding about missing values beforehand
```
