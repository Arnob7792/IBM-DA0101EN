```python
import pandas as pd
import numpy as np
df=pd.read_excel("C:\\Users\\ARNOB\\Desktop\\pricebinning.xlsx")
```

```python
#We will use pd.get_dummies() method
ftype=pd.get_dummies(df['fuel type'])
```

```python
#THIS WILL SAVE LIVES!!! MERGING TO EXISTING DATAFRAME to add the newfound columns to the right
df.merge(ftype,left_index=True,right_index=True)
```
