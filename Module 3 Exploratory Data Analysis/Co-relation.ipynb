```python
#Correlation is a statistical metric for
#measuring to what extent different variables are interdependent.
#In other words, when we look at two variables over time,
#if one variable changes how does this affect change in the other variable?
```

```python
#Co-relation doesnt imply causation
```

```python
#Lets test Co-relation by Positive Linear Relationship between engine size and price
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
df=pd.read_excel("C:\\Users\\ARNOB\\Desktop\\pricebinning.xlsx")
```

```python
sns.regplot(x='engine size',y='price',data=df)
plt.ylim(0,)
```

```python
# The regression line shows a positive linear co-relation between engine size and price
```

```python
#Now lets lok at the relationship between highway mpg and price
```

```python
sns.regplot(x='highway mpg',y='price',data=df)
plt.ylim(0,)
```

```python
# We can see a negative linear corelation from the regression line
```

```python
# Now we test a weak corelation
```

```python
sns.regplot(x='peak rpm',y='price',data=df)
plt.ylim(0,)
```

```python
# From the regression line we can see little to no corelation between the peak RPM and the price
```
