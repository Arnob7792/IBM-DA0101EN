```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
df=pd.read_excel("C:\\Users\\ARNOB\\Desktop\\pricebinning.xlsx")
```

```python
#Regression Plot
sns.regplot(x='highway mpg',y='price',data=df)
plt.ylim(0,)
```

```python
#Residual Plot: This is the plot created by subtracting the Target value from the predicted value
#Looking at the plot gives us some insight into our data.
#We expect to see the results to have zero mean.
#Distributed evenly around the x axis with similar variance; there is no curvature.
#This type of residual plot suggests a linear plot is appropriate.
```

```python
#If residuals are not uniformly distributed on both side of the predicted line LINEAR REGRESSION MAY BE INAPPROPRIATE
# RESIDUAL VALUES MAY APPEAR HYPERBOLIC OR SPREADING OUTWARDS, DO NOT USE LINEAR REGRESSION THEN
```

```python
#NOW WE WILL CREATE A RESIDUAL PLOT
```

```python
sns.residplot(df['highway mpg'],df['price']) #this is an alternative style to use seaborn
plt.ylim(0,)
#We see the residuals have a curvature which may imply that the linear assumption may be inappropriate
```

```python
#Distribution Plots
#A distribution plot counts the predicted value versus the actual value.
#These plots are extremely useful for visualizing models with more than one independent variable
```

```python
# A histogram is for discreet values
```

```python
#Now lets practice distribution plots for MLRs
```

```python
from sklearn.linear_model import LinearRegression #Import Linear Regression Model from Scikit-Learn
lm=LinearRegression() #Creating a empty Linear Regression objectusing the constructor
df=df.dropna()#It is essential to deal with any EmptyCells before trying regression.
#We can use the fill options with mean or something else
x=df[['highway mpg']]
y=df[['price']]
lm.fit(x,y)
Yhat=lm.predict(x)
```

```python
#THESE ARE THE CODES FOR THE DISTRIBUTION PLOT
ax1=sns.distplot(df['price'],hist=False,color='r',label="Actual Value")
sns.distplot(Yhat,hist=False,color='b',label='Fitted Values',ax=ax1)
#The actual values are used as a parameter.
#We want a distribution instead of a histogram so we want the hist parameter set to false.
#The color is red.
#The label is also included.
#The predicted values are included for the second plot;
#the rest of the parameters are set accordingly.
```

```python
#IF WE USE MORE VARIABLES, PREDICTION WILL BE MUCH CLOSER, LETS TRY
```

```python
z=df[['horsepower','curb weight','engine size','highway mpg']] # We will use 4 independent variables
#Then we train the model like before
lm.fit(z,df['price'])
Yhat=lm.predict(z)
```

```python
#DISTRIBUTION plOTTING FOR MULTIPLE VARIABLES
ax1=sns.distplot(df['price'],hist=False,color='r',label="Actual Value")
sns.distplot(Yhat,hist=False,color='b',label='Fitted Values',ax=ax1)
```
